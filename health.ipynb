```python
print("AI DIAGNOSTIC ASSISTANT ACTIVATED")
fever = input("Do you have fever? ")
cough = input("Do you have cough? ")
headache = input("Do you have headache?")
fatigue = input("Do you feel fatigue/tiredness?")
vomiting = input("do you have vomiting? ")
chestpain = input("do you have chest pain? ")
breathing = input("do you have breathing difficulty? ")
bodypain = input("do you have body pain? ")
thirst = input("do you feel excessive thirst? ")
urination = input("do you have frequent urination? ")
print("\nAnalyzing Symptoms...\n")
if fever and cough and breathing and fatigue:
    disease = "covidlike Infection"
    severity = "severe"
    suggestion = "consult a doctor immediately."
elif chestpain and breathing:
    disease = "heart-related Risk"
    severity = "severe"
    suggestion = "seek emergency medical care."
elif fever and bodypain and fatigue and headache:
    disease = "Flu"
    severity = "moderate"
    suggestion = "Take rest and stay hydrated."
elif cough and headache :
    disease = "common Cold"
    severity = "mild"
    suggestion = "Rest and drink warm fluid."

elif vomiting and fever:
    disease = "food Poisoning"
    severity = "moderate"
    suggestion = "Stay hydrated and eat light food."
elif headache and vomiting:
    disease = "migraine "
    severity = "Moderate "
    suggestion = "rest in a dark, quiet room."
elif thirst and urination and fatigue:
    disease = "Diabetes  Risk"
    severity = "moderate"
    suggestion = "Consult a doctor for blood sugar test."
else:
    disease = "No clear disease detected"
    severity = "Mildd"
    suggestion = "Maintainn a healthy lifestyle."
print("Possible Disease:", disease)
print("Severity:", severity)
print("Suggestion:", suggestion)
```
